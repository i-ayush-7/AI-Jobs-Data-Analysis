###### We will analyze this AI jobs dataset using Pandas to uncover hiring trends, salary patterns, and in-demand skills.
###### This includes exploring job counts, salary insights, company demand, skill frequency, and filtering roles.

```python
import pandas as pd
```

#### Load dataset

```python
df = pd.read_csv('Ai jobs dataset.csv', encoding= 'unicode_escape')
```

## EXPLORATORY DATA ANALYSIS

#### Q1. Count job listings per job title

```python
df['Job_Title'].value_counts()
```

#### Q2. Find average salary of each Job Title.

```python
df.groupby('Job_Title')['Salary'].mean().sort_values(ascending=False)
```

#### Q3. Top 5 highest job profiles.

```python
df.groupby('Job_Title')['Salary'].sum().sort_values(ascending=False).head(5).plot(kind='bar', figsize=(5,3), title='Top 5 highest job profiles.')
```

#### Q4. Find Average years of experience by Job Title.

```python
df.groupby('Job_Title')['Years_Experience'].mean().sort_values(ascending=False)
```

#### Q5. Find the Top 3 Job Pofiles by years of experience.

```python
df.groupby('Job_Title')['Years_Experience'].mean().sort_values(ascending=False).head(3).plot(kind='bar', figsize=(8,5), title='Top 3 Job Pofiles by years of experience')
```

#### Q6.Find Average salary by Education Level.

```python
df.groupby('Education_Level')['Salary'].mean().sort_values(ascending=False).plot(kind='bar', figsize=(8,5), title='Average salary by Education Level')
```

#### Q7. Find the top 5 Job Profiles with highest Technology Growth Factor.

```python
df.groupby('Job_Title')['Tech_Growth_Factor'].mean().sort_values(ascending=False).head(5).plot(kind='bar', figsize=(5,3), title='Top 5 Countries with highest GDP per Capita') 
```

#### Q8. Years of experience by Education level.

```python
df.groupby('Education_Level')['Years_Experience'].mean().sort_values(ascending=False).plot(kind='bar', figsize=(5,3), title='Years of experience by Education level')
```

#### Q9. Find Average Risk Categpry for the Technology Growth factor.

```python
df.groupby('Risk_Category')['Tech_Growth_Factor'].mean().sort_values(ascending=False).plot(kind='bar', figsize=(5,3), title='Average Risk Categpry for the Technology Growth factor') 
```

#### Q10. Find the top 10 Job Profiles with highest AI Exposure Index.

```python
df.groupby('Job_Title')['AI_Exposure_Index'].mean().sort_values(ascending=False).head(10).plot(kind='bar', figsize=(10,5), title='Top 5 Countries with highest GDP per Capita') 
```
